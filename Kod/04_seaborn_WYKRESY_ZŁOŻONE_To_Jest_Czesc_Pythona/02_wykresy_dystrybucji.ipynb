### HISTOGRAM = Distplot

```python
tips = sns.load_dataset('tips') # zaciagmu zbior danych 'tips'. tips to dataframe = dt
tips.head() #wyswietlamy kilka wiersz
```

```python
sns.displot(tips, x='tip') #tworzymy histogram dla kolumny z napiwkami - zobaczymy ile napiwków, po ile
```

```python
sns.displot(tips, x='tip', bins=30) # 'bins' = jaka granulacja/geśtopsc
```

### Wykresy relacyjne - Joinplot
# Joinplot = to wykres, który składa się z 3 różnych wykresów, gdzie w centralnym miejscu będzie wykres relacyjny, zaś u góry i po prawej będą histogramy kolumnowe


```python
_ = sns.jointplot(tips, x='total_bill', y='tip')
```

```python
_ = sns.jointplot(tips, x='total_bill', y='tip', hue='smoker') # relacja całkowitego rachunku i tipa, do smokera
# Tu dalej mamy relacje pomiędzy total bill, a  a jki tip + dodatkowo w rozbiciu na palacza lub nie palacza
```

### Relacja wszystkich kolumn - Pairplot = zbierze wszystkie wartości numeryczne, a następnie na ich bazie stworzy z nich wykresy relacyjne, czyli dostaniemy macierz wykresu, i wszyskie kolumny z wartościami numerycznymi ze sobą porównamy. Jedną metodą możemy zobaczyć na jednym wykresie relacje wszystkich kolumn, ktore zawierają dane numeryczne.

```python
_ = sns.pairplot(tips) # i tu daje cały zbiór, który został nazwany jako tips
```

```python
#ale można tez dodać zależność wartości numerycznych od innych-nienumerycznych -np. jak poniżej--> niebieskie czas "lunchu', pomar. 'dinner'
_= sns.pairplot(tips, hue = 'time') # ale za pomocą polecenia 'hue' możemy sobie jeszcze dodać rozbicie na jakiś element, w naszym przypdaku, rozbicie relacji wszystkich wartości numerycznych na porę dnia.
```
