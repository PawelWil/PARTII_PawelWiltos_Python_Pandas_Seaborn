```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') # tu sobie importuje te dane z bazy tips, w postaci tabeli
tips.head()
```

### Wykres z kategoriami(inaczej:Striplot). Zaś polecenie do wykresu z kategoriami to 'catplot'


```python
_ = sns.catplot(tips, x='day', y='total_bill') # jako kategorie obierzemy sobie dzien, i w zależności od dnia będziemy chcieli zobaczyć, jak wysokie były bille (czyli rachunki)
```

### SwarnPlot - czyli takie rozszerzenie, na bazie której widać, w ktorych miejscach(pozycjach) jest najwięcej wartości. I żeby to zrobić, muszę jako 'kind'=rodzaj wykresu dać 'swarm'

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm')
```

### Kolejny wykresy z kategoriami - gdzie dystrybucja danych jest za pomocą Boxplot-a (sporo informacji na temat danych!)
# Opis co ten wykres prezentuje: 1.Kreska po środku, pokazuje medianę, czyli średnią z całości, 2. wartości pod i nad kreską pokazują stosunek procentowy wartości poniżej mediany, do wartości powyżej mediany. 3. te wąsy pokazuje minima i ekstrema total billa, zaś te punkty u góry, pokazują wartości odstające, które wynosi okolo 3 %


```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='box') #i wiadomo, że za pomocą 'kind' zmieniłem pokazanie danych na typ box.
```

### Kolejny wykresy z kategoriami - WYKRES Słupkowy - BARchart, który pokazuje średnią dla danej kategorii, u nas to jest dla dnia, nie całkowitą wartość. Calkowita wartość będzie pokazana na poniższym wykresie 'count' + kreska pokazuje przedział pewności, gdzie z 95% pewnością srednia dla 'Total bill' będzie się wahała od 16 do 20


```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar')
```

### Kolejny wykresy z kategoriami - Ilość wystąpien - Countplot - ten wykres po prostu zlicza liści wystąpień (os y), w stosunku do kategorii, u nas day (os x), np. pokazuje ze w czwartek było okolo 62 rachunków, w tym zbiorze danych 'tips'

```python
_ = sns.catplot(tips, x='day', kind='count')
```
