```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') # tu sobie importuje te dane z bazy tips, w postaci tabeli
tips.head()
```

### Wykres z kategoriami(inaczej:Striplot). Za≈õ polecenie do wykresu z kategoriami to 'catplot'


```python
_ = sns.catplot(tips, x='day', y='total_bill') # jako kategorie obierzemy sobie dzien, i w zale≈ºno≈õci od dnia bƒôdziemy chcieli zobaczyƒá, jak wysokie by≈Çy bille (czyli rachunki)
```

### SwarnPlot - czyli takie rozszerzenie, na bazie kt√≥rej widaƒá, w ktorych miejscach(pozycjach) jest najwiƒôcej warto≈õci. I ≈ºeby to zrobiƒá, muszƒô jako 'kind'=rodzaj wykresu daƒá 'swarm'

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm')
```

### Kolejny wykresy z kategoriami - gdzie dystrybucja danych jest za pomocƒÖ Boxplot-a (sporo informacji na temat danych!)
# Opis co ten wykres prezentuje: 1.Kreska po ≈õrodku, pokazuje medianƒô, czyli ≈õredniƒÖ z ca≈Ço≈õci, 2. warto≈õci pod i nad kreskƒÖ pokazujƒÖ stosunek procentowy warto≈õci poni≈ºej mediany, do warto≈õci powy≈ºej mediany. 3. te wƒÖsy pokazuje minima i ekstrema total billa, za≈õ te punkty u g√≥ry, pokazujƒÖ warto≈õci odstajƒÖce, kt√≥re wynosi okolo 3 %


```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='box') #i wiadomo, ≈ºe za pomocƒÖ 'kind' zmieni≈Çem pokazanie danych na typ box.
```

### Kolejny wykresy z kategoriami - WYKRES S≈Çupkowy - BARchart, kt√≥ry pokazuje ≈õredniƒÖ dla danej kategorii, u nas to jest dla dnia, nie ca≈ÇkowitƒÖ warto≈õƒá. Calkowita warto≈õƒá bƒôdzie pokazana na poni≈ºszym wykresie 'count' + kreska pokazuje przedzia≈Ç pewno≈õci, gdzie z 95% pewno≈õciƒÖ srednia dla 'Total bill' bƒôdzie siƒô waha≈Ça od 16 do 20


```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar')
```

### Kolejny wykresy z kategoriami - Ilo≈õƒá wystƒÖpien - Countplot - ten wykres po prostu zlicza li≈õci wystƒÖpie≈Ñ (os y), w stosunku do kategorii, u nas day (os x), np. pokazuje ze w czwartek by≈Ço okolo 62 rachunk√≥w, w tym zbiorze danych 'tips'

```python
_ = sns.catplot(tips, x='day', kind='count')
```
