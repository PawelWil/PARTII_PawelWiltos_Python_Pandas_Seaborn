```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
_ = sns.regplot(tips,x='total_bill',y='tip') # wykres regresji, to ma reg przed plot.
# Regresja to metoda statystyczna suca okrelaniu zwizku pomidzy rnymi wielkociami i przewidywaniu=predykcji nieznanych wartoci jednych wielkoci na podstawie znanych wartoci innych.
# Ten wykres polega na pokazaniu wymaganych danych w postaci kropek (u nas to tip na osi y i total_bill na osi x), a nastpnie na dopasowaniu linii do najwikszego skupienia, a jednoczenie przewidywaniu=predykcji, na bazie tego zachowania jak si to bdzie zachowywa pniej.
# Jak wida oprcz cigej linii, mamy rwnie zacieniony obszar, ktory jakby poszerza moliwy zakres wystpienia wartoci. Im wikszy ten cie, tym wiekszy rozrzut i mniejsza powtarzalno, czyli predykcja prawdopodobnie bdzie obarczona duym bdem.
```
