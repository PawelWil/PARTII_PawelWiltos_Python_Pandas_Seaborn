### Wykres relacyjny, typu 'rozrzutu' = 'scatter'

```python
tips = sns.load_dataset('tips') # tu sobie importuje te dane z wewnętrznej bazy Seabron, w ktorej ten zbiór występuje pod nazwą 'tips' - i dlatego nazwaliśmy go 'tips', żeby wiedział do czega ma się odnosić
# w postaci tabeli, a za pomocą head biore tylko pare górnych wartoscoi
tips.head() # wyswietlamy za pomocą 'head' kilka pierwszych wierszy
```

```python
# sns.relplot(baza danych, x=dana wykresu x, y=dana wykresu y) - za pomocą tej funkcji rysujemy w seaborn wykresy
sns.relplot(tips, x='total_bill', y = 'tip' ) # wykresy relacyjne za pomocą 'relplot'. tips=baza danych(w naszym przypadku to dane, z wewnętrznej bazy Seaborn), a teraz dajemy dane dla osi x(2 parametr) oraz dla osi y (3parametr)
tips.head() # wyswietlamy za pomocą 'head' kilka pierwszych wierszy
```

```python
_ = sns.relplot(tips, x='total_bill', y='tip', hue='sex') # 'hue' daje dodatkowe info + legendę
```

```python
sns.relplot(tips, x='total_bill', y='tip', hue='sex', size='size')#tu za pomocą paramtru 'size', możemy
# pokazać wielkość/natężenie dladanych parametrów, w naszym przypadku to są różnej wielkości kropki
```

### Wykres relacyjny - typ liniowy

```python
ice = sns.load_dataset('seaice') #importujemy dane, żeby z nich moc w dalszej czesci twoerzenia wykresów skorzystac. Przypisuje to równiez do zmiennej 'ice', która to nazwa jest dalej używana
ice.head()
```

```python
ice.info()
```

```python
_ = sns.relplot(ice[ice['Date'].dt.year == 2018], x='Date', y='Extent', kind='line') # parametr 'kind' zmienia typ wykresy, np. teraz ze skatter, na line
```

```python
ice['mm-dd'] = ice['Date'].dt.strftime('%m-%d')  #teraz dodamy nową kolumnę do naszego DataFrame=dt, któa bedzie reprezentowało miesiąc i dzien - zastąpimy kolumnę 'Date', na'mm-dd'
```

```python
ice.head() # polecenie 'head' pokazuje nam tylko kilka pierwszych wiersz
```

```python
ice_2000_2018 = ice[(ice['Date'].dt.year > 1999) & (ice['Date'].dt.year < 2019)] # tu dodaliśmy zmienną, która
#definiuje okres, tak żeby potem nie przepisywac wszystkiego, tylko wrzucić zmienną
```

```python
_ = sns.relplot(ice_2000_2018, x='mm-dd', y='Extent', kind='line')
```
